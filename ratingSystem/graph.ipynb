```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('only.csv')

final_rating = df['finalrating']
initial_rating = df['initialrating']

plt.figure(figsize=(8, 6))
plt.scatter(initial_rating, final_rating, color='blue', label='Initial Rating', alpha=0.7)
plt.scatter(initial_rating, initial_rating, color='red', label='Final Rating', alpha=0.7)
plt.title('Initial Rating vs Final Rating')
plt.xlabel('Initial Rating')
plt.ylabel('Final Rating')
plt.legend()
plt.grid(True)
plt.show()
```
